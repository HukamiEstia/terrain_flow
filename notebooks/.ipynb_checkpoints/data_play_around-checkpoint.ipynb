```python
from matplotlib import pyplot as plt
from osgeo import gdal
import numpy as np
import plotly.graph_objects as go
from PIL import Image
```

```python
ds = gdal.Open('/home/data/n15_e022_1arc_v3.dt2')
band_count = ds.RasterCount
print(band_count)
bnd = ds.GetRasterBand(1) #1 based
arr = bnd.ReadAsArray()
plt.imshow(arr)
plt.colorbar()
```

```python
arr2 = arr[1000:1050,200:250]
plt.imshow(arr2)
plt.colorbar()
```

```python
arr2 = np.asarray(arr2)
im = Image.fromarray(arr2)
im.save('/home/data/test1.tif')
```

```python
test = gdal.Open('/home/data/test1.tif')
band_count = test.RasterCount
band_count
bnd = test.GetRasterBand(1) #1 based
test = bnd.ReadAsArray()
plt.imshow(test, cmap='gray')
plt.colorbar()
```

```python
select = np.zeros((600,600,4))
n=0
for i in range(600):
    for j in range(600):
        if test[i][j] <= 650:
            n+=1
            select[i][j][0] = 1
            select[i][j][3] = 0.5
fig, ax = plt.subplots()
ax.imshow(test, cmap='gray')
ax.imshow(select)
n
```

```python
fig = go.Figure(data=[go.Surface(x=[30*i for i in range(len(test))], y=[30*i for i in range(len(test))], z=test, colorscale="viridis")])
fig.update_layout(title='DEM map test', autosize=True,
                  scene = {"xaxis":{"visible":False}, "yaxis":{"visible":False}, "zaxis":{"nticks":4, "range":[0,1500], "visible":False}},
                  margin={"l":0, "r":0, "b":0, "t":0})
fig.show()
```

```python
fig = go.Figure(data=[go.Surface(x=[30*i for i in range(len(test))],
                                 y=[30*i for i in range(len(test))],
                                 z=test,
                                 colorscale="Greens")])
fig.update_layout(title='DEM map test', autosize=True,
                  margin={"l":0, "r":0, "b":0, "t":0},
                  scene_aspectmode='data',
                  scene={"xaxis":{"visible":False},
                         "yaxis":{"visible":False},
                         "zaxis":{"visible":False}}
                 )
fig.show()
```
